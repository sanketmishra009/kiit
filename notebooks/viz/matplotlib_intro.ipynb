Scatter


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,20)
colors = ['green'] * (x.size -10)
colors+=['red']*10

y = x

plt.scatter(x , y , s = 100 , c = colors)
```

```python
plt.scatter(x[:12] , y[:12] , c= 'green' , label= 'Short Students')
plt.scatter(x[12:] , y[12:] , c='red' , label = 'Tall Students')
plt.legend(loc = 4 , title = 'Legend'  , frameon = False)
plt.gca().get_children()
legend = plt.gca().get_children()[-2]
# legend.get_children()[0].get_children()[1].get_children()[0].get_children()
from matplotlib.artist import Artist

def get_rc(art , depth = 0):
    if isinstance(art , Artist):
        print(" "* depth + str(art))
        for child in art.get_children():
            get_rc(child , depth + 2)


get_rc(legend)
```

Line Graph

```python
linear = np.arange(0,20)
quadratic  = (linear**2 )+ 1
```

```python
plt.plot(range(len(linear)) , linear , '-o' , quadratic , '-o')
plt.gca().fill_between(range(len(linear)) , linear , quadratic , facecolor='blue' , alpha = 0.25)
```
